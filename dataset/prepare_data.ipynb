```python
def load_channel_detail(path):
    output_list = []
    with open(path, "rb") as f:
        for line in f:
            output_list.append(json.loads(line))
    return output_list
```

```python
json_l=load_channel_detail("json-channelDetail.json")
```

```python
# json_l
```

```python
def extract_channel_detail(json_list):
    channel_list=[]
    desc_list=[]
    for item in json_list:
        channel_list.append(item.get("name", "null"))
        desc_list.append(item.get("desc", "null"))
    df = pd.DataFrame.from_dict({
        "channel": channel_list,
        "channel_desc": desc_list,
    })
    return df
```

```python
ch_mi=extract_channel_detail(json_list=json_l)
```

```python
json_l_tf=load_channel_detail("json-triggerList.json")
```

```python
def extract_function_detail(json_list, split):
    function_list=[]
    desc_list=[]
    channel_list=[]
    field_list=[]
    for item in json_list:
        function_list.append(item.get(f"{split}Title", "null"))
        desc_list.append(item.get(f"{split}Desc", "null"))
        channel_list.append(item.get(f"{split}ChannelName", "null"))
        field_list.append(item.get(f"{split}FieldList", "null"))
    df = pd.DataFrame.from_dict({
        "channel": channel_list,
        "function": function_list,
        "desc": desc_list,
        "field": field_list
    })
    return df
```
